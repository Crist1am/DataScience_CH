Cada año que pasa, las personas tienen más posibilidades de viajar a otros países y uno de los más atractivos para los países sudamericanos es España, principalmente por el idioma. España se caracteriza por tener ciudades muy turísticas en el verano europeo, como Barcelona, Madrid, la isla de Mallorca, entre otras. Existen muchas aplicaciones para alojar en distintas ciudades del mundo, dentro de las cuales su precio depende de muchos factores, como cercanía a atractivos turísticos, los servicios que ofrecen, la capacidad de personas admitidas, etc.
<b>Este proyecto busca generar un modelo que permita estimar el precio de renta de alojamientos turísticos, ya sea de una habitación privada o de la renta de una casa/apartamento en la ciudad de Barcelona</b>, comprendiendo las relaciones entre las características de los alojamientos y sus respectivas rentas, con el fin de ayudar a los anfitriones a ajustar sus valores de renta y a los que quieran incursionar en esta área de arrendar alojamientos turísticos tomando la mejor decisión.

><h3> Identificar valores atipicos </h3>

Para identificar los valores atipicos en cada columna con valores numericos se creó una función que, mediante el metodo de los cuartiles, ya que este metodo es robusto y se puede utilizar en datos que no siguen una distribución normal (como es en este caso).
<br>

```python
def filtro_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    filtro = (df[col] < lower) | (df[col] > upper)
    df[col] = df[col].where(~filtro, np.nan)
    return df
```

De esta manera, utilizamos la función con cada columna del df para identificar los <i>outliers</i> y eliminar el dato (y no 'castigar' la fila completa), así mantenemos el tamaño del df.

Para esto, se utilizó un ciclo for que iteró en cada columna que presentó nulos y remplazaba los NaN por la mediana del conjunto de datos.

```python
col_con_nulos = ['price', 'accommodates', "cant_comodidades",'bathrooms', 'bedrooms', 'beds']

for i in col_con_nulos:
  df[i].fillna(df[i].median(), inplace = True)
```
